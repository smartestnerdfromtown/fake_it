```python
import pandas as pd
from name import (
    generate_name,
    generate_surname,
    generate_fullname
)
from mail import generate_email
from location import generate_country
```

```python
samples = 1000000


names = [generate_name() for _ in range(samples)]
surnames = [generate_surname() for _ in range(samples)]
emails = [generate_email(names[i], surnames[i], random_domain=True) 
                                                for i in range(samples)]
countries = [generate_country() for _ in range(samples)]



df = pd.DataFrame({
    "name": names,
    "surname": surnames,
    "email": emails,
    "country": countries,
})
```

```python
df
```

```python
(
    df
    .groupby(by="country")
    .agg(number_of_users=("country", "count"))
    .sort_values(by="number_of_users", ascending=False)
)
```
